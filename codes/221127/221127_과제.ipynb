### date_range
- 주어진 기간의 날짜(구간)를 DatetimeIndex 타입으로 반환
- pd.date_range() 형태로 사용
- 매개변수: start, end, freq 등

```python
pd.date_range("2021-1-1", "2021-1-15") # freq의 기본값: D(일)
# pd.date_range(start="2021-1-1", end="2021-1-15")
```

```python
pd.date_range("2021-1-1", "2021-1-15", freq="B") # B(평일)
```

```python
pd.date_range("2022-1-13", "2023-1-27", freq="W") # W(주)
```

```python
pd.date_range("2002-1-1", "2022-5-31", freq="A") # A(연도별 마지막 날)
```

```python
pd.date_range("2020-1-1", "2023-12-31", freq="Q") # Q(분기별 마지막 날)
```

### ffill, bfill
- 결측치가 있는 데이터를 앞/뒤 데이터로 대체
- 시계열 데이터를 더 잘게 쪼갤 때 유용
- df.ffill() 또는 df.bfill() 형태로 사용
- 매개변수: axis

```python
df = pd.DataFrame({"A":[1, 2, 3, 4, 5], 
                   "B":[2.18, 5.93, 6.51, 1.13, 7.28], 
                   "C":[398, 970, 269, 481, 477], 
                   "D":[12, 51, 31, 6, 34], 
                   "E":[10, 212, 15, 56, 710]})
```

```python
time_id = pd.to_datetime(pd.Series(["2022-10-31", "2022-11-5", 
                                    "2022-11-7", "2022-12-20", "2023-1-13"]))
```

```python
df["date"] = time_id
```

```python
df = df.set_index("date")
```

```python
df.resample("W").mean() # 없는 데이터는 NaN으로 표시
```

```python
df.resample("W").mean().ffill() # 결측치를 앞 데이터로 변경
```

```python
df.resample("W").mean().bfill() # 결측치를 뒷 데이터로 변경
```

### first, last
- 각 그룹의 첫/마지막 데이터 반환
- df.first() 또는 df.last() 형태로 사용
- 시계열 인덱스 데이터에 사용 시 (sort 후) offeset 변수 필요

```python
crime = pd.read_csv("crime.csv")
```

```python
crime["REPORTED_DATE"] = pd.to_datetime(crime["REPORTED_DATE"])
```

```python
crime = crime.set_index("REPORTED_DATE").sort_index()
```

```python
crime.first("1D") # 첫 날(2012-01-02) 데이터만 반환
```

```python
crime.first("1S") # 1초 단위(2012-01-02 00:06:00)로 데이터만 반환
```

```python
crime.first("3A") # 앞에서부터 3년 데이터만 반환
```

```python
crime.last("2W") # 뒤에서 2주 데이터만 반환
```

```python
crime.last("5D") # 뒤에서 5일 데이터만 반환
```

```python
crime.last("5B") # 뒤에서 5일(평일) 데이터만 반환
```

### offset - DateOffset
- 시계열 타입 데이터에 더하고 뺄 수 있는 오프셋을 반환
- pd.DateOffset() 형태로 사용
- 매개변수: days, months, years 등

```python
date1 = pd.to_datetime("2001-07-30")
```

```python
date1
```

```python
y1 = pd.DateOffset(years=1)
```

```python
date1 + y1 # 2001-07-30에 1년을 더한 값이 출력
```

```python
m1 = pd.DateOffset(months=5)
```

```python
date1 + m1 # 2001-07-30에 5달을 더한 값이 출력
```

```python
d0 = pd.DateOffset(years=3, months=2, days=15, 
                   hours=4, minutes=30, seconds=47)
```

```python
date1 + d0
```

## 필수과제2
- 시계열데이터를 가지고 요일을 추출하는 것을 구글링해서 컬럼을 만들어주세요!
- 요일컬럼 만들어서 공유!

```python
# 인덱스(시계열 데이터) 추출
crime_date = crime.index.to_series().reset_index(drop=True)
```

```python
crime_date.to_frame()
```

```python
crime_weekday = crime_date.dt.dayofweek # dayofweek: 날짜의 요일을 알려줌
```

```python
crime_weekday # 요일 정보를 추출해서 별도로 저장
```

```python
crime_weekday.value_counts().sort_index().to_frame() # 데이터의 요일 분포를 나타냄
```

#### 숫자별 해당되는 요일
- 0 : 월요일(Monday)
- 1 : 화요일(Tuesday)
- 2 : 수요일(Wednesday)
- 3 : 목요일(Thursday)
- 4 : 금요일(Friday)
- 5 : 토요일(Saturday)
- 6 : 일요일(Sunday)

```python
# 숫자로 된 요일 정보를 문자로 변환
weekday_name = {0:"Monday", 1:"Tuesday", 2:"Wednesday", 
                3:"Thursday", 4:"Friday", 5:"Saturday", 6:"Sunday"}
```

```python
crime_weekday2 = crime_weekday.replace(weekday_name).to_frame()
```

```python
crime_weekday2.rename(columns={"REPORTED_DATE":"weekdays"}, inplace=True)
```

```python
crime_weekday2
```

```python
# 변환시킨 요일 데이터를 crime 데이터프레임의 컬럼으로 추가
crime["weekdays"] = crime_weekday2["weekdays"].values
```

```python
crime
```

```python
crime["weekdays"].sample(3).to_frame() # 요일 컬럼을 바르게 만들었음을 확인할 수 있음
```
