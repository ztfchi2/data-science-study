```python
import pandas as pd
```

```python
df_train = pd.read_csv('train.csv')
df_train
```

```python
# start_turn_restricted, end_turn_restricted
# 둘 다 없음, 있음의 답변을 가지므로 0,1로 바꾸어줌
df_train['start_turn_restricted'] = df_train['start_turn_restricted'].apply(lambda x: 0 if x=='없음' else 1)
df_train['start_turn_restricted']
```

```python
df_train['end_turn_restricted'] = df_train['end_turn_restricted'].apply(lambda x: 0 if x=='없음' else 1)
df_train['end_turn_restricted']
```

```python
# 'id','vehicle_restricted','height_restricted','start_node_name','end_node_name','road_name'
# id의 경우 의미가 없어서 지움
# 'vehicle_restricted','height_restricted': 모든 값이 0이라서 의미가 없음
# 'start_node_name','end_node_name': 각각 start_latitude, start_longitude / end_latitude, end_longitude로 표현가능하므로 지움.
# 'road_name': 결과값의 종류가 너무 많아서 지움
df_train.drop(['id','vehicle_restricted','height_restricted','start_node_name','end_node_name','road_name'],axis=1,inplace=True)
```

```python
# 날짜 데이터를 분기별로 나누어서 전치리 진행
df_train['base_date'] = df_train['base_date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
```

```python
df_train['base_quarter'] = df_train['base_date'].apply(lambda x: 1 if x.month<=3 else 2 if x.month<=6 else 3 if x.month<=9 else 4)
```

```python
# day_of_week, connect_code, weight_restricted, road_type, base_hour, base_quarter, road_rating
# 전부 원 핫 인코딩 처리
week_day = pd.get_dummies(df_train['day_of_week'])
df_train = pd.concat([df_train,week_day],axis=1)
```

```python
connect = pd.get_dummies(df_train['connect_code'])
df_train = pd.concat([df_train,connect],axis=1)
```

```python
weight = pd.get_dummies(df_train['weight_restricted'])
df_train = pd.concat([df_train,weight],axis=1)
```

```python
road_type = pd.get_dummies(df_train['road_type'])
df_train = pd.concat([df_train, road_type],axis=1)
```

```python
hour = pd.get_dummies(df_train['base_hour'])
df_train = pd.concat([df_train,hour],axis=1)
```

```python
quarter = pd.get_dummies(df_train['base_quarter'])
df_train = pd.concat([df_train,quarter],axis=1)
```

```python
rate = pd.get_dummies(df_train['road_rating'])
df_train = pd.concat([df_train,rate],axis=1)
df_train
```

```python
df_train.drop(['base_date','day_of_week','connect_code','weight_restricted','road_type','base_hour','base_quarter','road_rating'], axis=1,inplace=True)
```

```python
df_train.columns = [           'lane_count',          'multi_linked',
         'maximum_speed_limit',        'start_latitude',
             'start_longitude', 'start_turn_restricted',
                'end_latitude',         'end_longitude',
         'end_turn_restricted',                'target',
                           '금',                     '목',
                           '수',                     '월',
                           '일',                     '토',
                           '화',                       'code_0',
                           'code_103',                    'weight_0',
                       'weight_32400',                 'weight_43200',
                       'weight_50000',                       'road_0',
                             'road_3',                       'road_0',
                             'hour_1',                       'hour_2',
                             'hour_3',                       'hour_4',
                             'hour_5',                       'hour_6',
                             'hour_7',                       'hour_8',
                             'hour_9',                      'hour_10',
                            'hour_11',                      'hour_12',
                            'hour_13',                      'hour_14',
                            'hour_15',                      'hour_16',
                            'hour_17',                      'hour_18',
                            'hour_19',                      'hour_20',
                            'hour_21',                      'hour_22',
                            'hour_23',                       'quarter_1',
                             'quarter_2',                       'quarter_3',
                             'quarter_4',                     'rating_103',
                           'rating_106',                     'rating_107']
```

```python
# 위도와 경도 표준화 진행
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df_train[['start_latitude','start_longitude']])
scaled = scaler.transform(df_train[['start_latitude','start_longitude']])
```

```python
scaled
```

```python
df_train[['start_latitude','start_longitude']] = scaled
```
