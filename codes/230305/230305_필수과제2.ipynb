### 전체 데이터 전처리 계획(확정X)
1. base_date: 시계열 데이터로 변환 후 연/월/일 나누기
2. day_of_week: 정수형 데이터로 전환, 각 날에 대해 원-핫 인코딩, 주말 변수로 원-핫 인코딩
3. base_hour: 시간대 데이터, 출퇴근 시간대 나누어서 원-핫 인코딩
4. lane_count: 차로수 데이터 (건드리지 않음)
5. road_rating: 도로등급 데이터, 원-핫 인코딩으로 변경?
6. road_name: 도로명 텍스트 전처리 진행, '-'는 결측치로 생각, 각각에 대해 원-핫 인코딩을 하되 일반국도, 지방도 변수도 추가, -----
7. multi_linked: 중용구간 여부 데이터, 그대로 둠
8. connect_code: 연결로 코드, 0-1로 레이블 인코딩
9. maximum_speed_limit: 정수형 데이터로 전환, ***
10. vehicle_restricted: 전부 0.0인 데이터, -----
11. weight_restricted: 숫자 크기에 따라 레이블 인코딩?
12. height_restricted: 전부 0.0인 데이터, -----
13. road_type: 도로유형 데이터, 0-1로 레이블 인코딩

14. start_node_name: 시작지점명 데이터, 종류가 다양함, -----
15. start_latitude: 시작지점 위도 데이터, 정규화 진행?
16. start_longitude: 시작지점 경도 데이터, 정규화 진행?
17. start_turn_restricted: 시작지점 회전제한 유무, 0-1로 레이블 인코딩

18. end_node_name: 시작지점과 같이 진행, -----
19. end_latitude: 시작지점과 같이 진행
20. end_longitude: 시작지점과 같이 진행
21. end_turn_restricted: 시작지점과 같이 진행

22. target: y로 사용할 데이터

### 시계열 데이터 관련
- 1, 2, 3번 열 전처리

```python
train_df["base_date"]
```

```python
from datetime import datetime
```

```python
# base_date 변수를 시계열 데이터로 변환
train_df["base_date"] = pd.to_datetime(train_df["base_date"], format="%Y%m%d")

test_df["base_date"] = pd.to_datetime(test_df["base_date"], format="%Y%m%d")
```

```python
# year, month, quarter 변수 생성
train_df["year_dt"] = train_df["base_date"].dt.year
train_df["month_dt"] = train_df["base_date"].dt.month
train_df["quarter_dt"] = train_df["base_date"].dt.quarter

test_df["year_dt"] = test_df["base_date"].dt.year
test_df["month_dt"] = test_df["base_date"].dt.month
test_df["quarter_dt"] = test_df["base_date"].dt.quarter
```

```python
# quarter 원-핫 인코딩 진행
train_df["quarter_dt"].replace({1:"Q1", 2:"Q2", 3:"Q3", 4:"Q4"}, inplace=True)
oh_q1 = pd.get_dummies(train_df["quarter_dt"])
train_df = pd.concat([train_df, oh_q1], axis=1)

test_df["quarter_dt"].replace({1:"Q1", 2:"Q2", 3:"Q3", 4:"Q4"}, inplace=True)
oh_q2 = pd.get_dummies(test_df["quarter_dt"])
test_df = pd.concat([test_df, oh_q2], axis=1)
```

```python
# year 원-핫 인코딩 진행
oh_y1 = pd.get_dummies(train_df["year_dt"])
oh_y1 = oh_y1.rename(columns={2021:"2021", 2022:"2022"})
train_df = pd.concat([train_df, oh_y1], axis=1)

oh_y2 = pd.get_dummies(test_df["year_dt"])
oh_y2 = oh_y2.rename(columns={2021:"2021", 2022:"2022"})
test_df = pd.concat([test_df, oh_y2], axis=1)
```

```python
train_df.year_dt.value_counts().to_frame()
```

```python
train_df["base_date"].dt.weekday
```

```python
train_df["day_of_week"]
# base_date의 weekday와 같은 값
```

```python
# day_of_week 변수값을 영어로 변경
train_df["day_of_week"].replace({"월":"Mon", "화":"Tue", "수":"Wed", "목":"Thu", 
                                 "금":"Fri", "토":"Sat", "일":"Sun"}, inplace=True)

test_df["day_of_week"].replace({"월":"Mon", "화":"Tue", "수":"Wed", "목":"Thu", 
                                "금":"Fri", "토":"Sat", "일":"Sun"}, inplace=True)
```

```python
# 각 요일별 원-핫 인코딩 진행
oh_w1 = pd.get_dummies(train_df["day_of_week"])
train_df = pd.concat([train_df, oh_w1], axis=1)

oh_w2 = pd.get_dummies(test_df["day_of_week"])
test_df = pd.concat([test_df, oh_w2], axis=1)
```

```python
# business_day 변수 생성
train_df["business_day"] = train_df["day_of_week"].replace({"Mon":1, "Tue":1, "Wed":1, "Thu":1, "Fri":1, "Sat":0, "Sun":0})

test_df["business_day"] = test_df["day_of_week"].replace({"Mon":1, "Tue":1, "Wed":1, "Thu":1, "Fri":1, "Sat":0, "Sun":0})
```

```python
train_df["base_hour"]
# 6-9시를 출근 시간, 18-21시를 퇴근 시간으로 설정
```

```python
# rush_hour 변수 생성
train_df["rush_hour"] = np.where(((train_df["base_hour"] >= 6) & (train_df["base_hour"] <= 9)) | ((train_df["base_hour"] >= 18) & (train_df["base_hour"] <= 21)), 1, 0)

test_df["rush_hour"] = np.where(((test_df["base_hour"] >= 6) & (test_df["base_hour"] <= 9)) | ((test_df["base_hour"] >= 18) & (test_df["base_hour"] <= 21)), 1, 0)
```
