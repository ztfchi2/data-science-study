### 모델링을 들어가기 전 하는 과정
- 전처리 -> 모델링
- 전처리, 피처엔지니어링, 스케일링, 인코딩 등 전처리 완료 후에
- 모델링 전 train, val, test 데이터를 분리한다.
- 왜 우리는 데이터를 나누는가?

- 모델링 -> 모델링은 지도학습 분류, 예측 -> 분류예측하고 끝, 분석가는 결국 분류와 예측한 결과를 가지고 -> solution이 나와야한다!
- 솔루션이지만 -> 예측한 모델링을 단순하게 한 번 하고 끝나는게 아니라 결국 일반화하기 위해서 진행한다.
- 일반화를 하기 위해서는 과적합을 피해야 한다.
- ex) 수능, 모의고사 비교
- 데이터 학습을 과하게 해서 실제 train 데이터에는 좋은 결과가 나오지만 test는 좋지 못한 결과가 나오게 된다.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris_data = iris.data
```

```python
iris_data
```

```python
# 잘 모르시면 Iris 데이터 구글링하면 다 나옵니다!
iris_label = iris.target
```

```python
iris_label
```

```python
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['y'] = iris_label
```

```python
iris_df
```

### 학습용 데이터, 테스트용 데이터 분리
- 학습 데이터로 학습된 모델이 얼마나 뛰어난 성능을 가지고 있는지 평가하려면 test 데이터가 필요하다.
    - 사이킷런에서 제공하는 train_test_split() API 제공
    - test(데이터 분할 가능), test_size 조정 가능하다. 전체 데이터를 내가 직접 비중을 나눠서 train, test로 나눌 수 있다.

```python
iris_df
```

```python
# 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.3, random_state=111)
# test_size : 전체 데이터에서 비중을 말하는 것, random_state : 섞는 개념으로 이해(시드값 지정)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
df_clf = DecisionTreeClassifier(random_state=111) # 결정트리 사용
```

```python
df_clf.fit(X_train, y_train) # fit 안에 우리가 넣은 건 X_train, y_train : x 데이터와 이에 대응하는 y값
```

```python
# 예측을 진행하자! 학습이 완료되었다.
pred = df_clf.predict(X_test)
```

```python
pred # 정답을 출력 (X_test 값을 학습해서 y값을 예측했다!)
```

```python
# 정확도를 봐야 알지!
from sklearn.metrics import accuracy_score
```

```python
print("DT 예측 정확도: {0:.3f}".format(accuracy_score(y_test, pred)))
# y_test 기존의 정답, pred 예측값
```
