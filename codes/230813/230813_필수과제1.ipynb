```python
import pandas as pd
import numpy as np
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
credit = pd.read_csv("creditcard.csv")
```

```python
# Class를 문자형으로 변경
credit["Class"] = credit["Class"].astype("str")
```

```python
credit.Class.value_counts()
```

```python
y = np.array(credit["Class"])
X = credit.drop(["Class"], axis=1, inplace=False)
```

### 기본값 voting 파라미터

```python
# voting에 필요한 모델을 학습하자
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=3)
dt_clf = DecisionTreeClassifier()

# (hard) voting 불러오기
vo_clf = VotingClassifier(estimators=[("LR",lr_clf),("KNN",knn_clf),("DT",dt_clf)])

feature = X
feature = np.array(feature)
X_train, X_test, y_train, y_test = train_test_split(feature, y, test_size=0.2, random_state=111)

vo_clf.fit(X_train, y_train)
pred = vo_clf.predict(X_test)
print(accuracy_score(y_test, pred))
```

```python
# voting에 필요한 모델을 학습하자
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier(n_neighbors=3)
dt_clf = DecisionTreeClassifier()

# (soft) voting 불러오기
vo_clf = VotingClassifier(estimators=[("LR",lr_clf),("KNN",knn_clf),("DT",dt_clf)], 
                          voting="soft")

feature = X
feature = np.array(feature)
X_train, X_test, y_train, y_test = train_test_split(feature, y, test_size=0.2, random_state=111)

vo_clf.fit(X_train, y_train)
pred = vo_clf.predict(X_test)
print(accuracy_score(y_test, pred))
```

- 기본값일 때 hard voting보다 soft voting이 성능이 더 높으나, 차이가 유의미하지는 않다.

### 최종 voting 파라미터

```python
# voting에 필요한 모델을 학습하자
lr_clf = LogisticRegression(penalty="l2", C=0.1, max_iter=500)
knn_clf = KNeighborsClassifier(n_neighbors=2)
dt_clf = DecisionTreeClassifier(max_depth=5, min_samples_split=4)

# (hard) voting 불러오기
vo_clf = VotingClassifier(estimators=[("LR",lr_clf),("KNN",knn_clf),("DT",dt_clf)])

feature = X
feature = np.array(feature)
X_train, X_test, y_train, y_test = train_test_split(feature, y, test_size=0.2, random_state=111)

vo_clf.fit(X_train, y_train)
pred = vo_clf.predict(X_test)
print(accuracy_score(y_test, pred))
```

```python
# voting에 필요한 모델을 학습하자
lr_clf = LogisticRegression(penalty="l2", C=0.1, max_iter=500)
knn_clf = KNeighborsClassifier(n_neighbors=2)
dt_clf = DecisionTreeClassifier(max_depth=5, min_samples_split=4)

# (soft) voting 불러오기
vo_clf = VotingClassifier(estimators=[("LR",lr_clf),("KNN",knn_clf),("DT",dt_clf)], 
                          voting="soft")

feature = X
feature = np.array(feature)
X_train, X_test, y_train, y_test = train_test_split(feature, y, test_size=0.2, random_state=111)

vo_clf.fit(X_train, y_train)
pred = vo_clf.predict(X_test)
print(accuracy_score(y_test, pred))
```
