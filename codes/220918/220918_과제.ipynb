```python
from matplotlib import pyplot as plt
```

```python
l1 = [1,2,5,4,3]
l2 = [11,22,92,58,30]
l3 = [0,1,1,0,1]
```

### scatter

```python
plt.scatter(l1,l2) # 기본
```

```python
plt.scatter(l1,l2, c=l3) # c: 점(마커)의 색 지정
```

```python
plt.scatter(l1,l2, c=l3, edgecolors="red") # edgecolors: 마커의 테두리 색 지정
```

```python
plt.scatter(l1,l2, c="white", edgecolors="blue", 
           linewidth=15) # linewidth: 마커의 테두리 두께를 바꿈
```

### plot

```python
import numpy as np
```

```python
plt.plot(np.random.normal(0,100,1000).cumsum()) # 기본
```

```python
plt.plot(np.random.normal(0,100,100).cumsum(), 
         marker="x", color="green") # 초록색 x형 마커
```

```python
plt.plot(np.random.normal(0,100,1000).cumsum(), "gx") # 이런 식으로도 가능
```

```python
plt.plot([1,2,3,4,5],[2,4,6,8,10]) # 기본
```

```python
plt.plot([1,2,3,4,5],[2,4,6,8,10], 
         marker='.', markersize=10, color="blue", linewidth=2, linestyle="dashed")
```

```python
plt.plot([1,2,3,4,5],[2,4,6,8,10], "r|--", linewidth=2, markersize=15)
```

```python
plt.plot([1,2,3,4,5],[2,4,6,8,10], "m*--", # markeredgecolor: 마커 테두리 색 지정
         linewidth=2, markersize=15, markeredgecolor="black")
```

### title, xlab, ylab

```python
# 한글 사용 코드
from matplotlib import font_manager, rc, rcParams

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family = font_name)
    
rcParams['axes.unicode_minus']=False
```

```python
# 한글 폰트 적용 확인
plt.plot(range(1,11), range(3,31,3), marker='o')
plt.title("선형 그래프")
plt.xlabel("x축")
plt.ylabel("y축")
```

```python
plt.plot(range(1,11), range(3,31,3), marker='o', color="black")
plt.title("선형 그래프", fontsize=30) # fontsize: 폰트 크기 지정
plt.xlabel("x축", fontsize=15, color="green") # 축 이름도 색 변경 가능
plt.ylabel("y축", fontsize=15, rotation=45) # 축 이름 회전 가능
```

```python
plt.plot(range(1,11), range(3,31,3), marker='o', color="black")
plt.title("linear graph", fontsize=25, pad=15, # pad: 타이틀 여백 설정
          color="blue", backgroundcolor="skyblue") # backgroundcolor: 배경 색 지정
plt.xlabel("x", fontsize=15, loc="right", alpha=0.25) # 축 위치, 투명도 변경 가능
plt.ylabel("y", fontsize=15, labelpad=10) # labelpad: 축 여백 설정
```

### xlim, ylim, xticks, yticks

```python
plt.plot(l1, l2, marker='+', color="yellow", 
         markersize=10, markeredgecolor="black", linewidth=2)
```

```python
plt.plot(l1, l2, marker='+', color="yellow", 
         markersize=10, markeredgecolor="black", linewidth=2)
plt.xlim(left=2, right=4) # plt.xlim(3,8)과 동일
plt.ylim(bottom=20, top=80) # plt.ylim(10,27)과 동일
```

```python
plt.plot(l1, l2, marker='+', color="yellow", 
         markersize=10, markeredgecolor="black", linewidth=2)
plt.xticks(range(1,6), ['A','B','C','D','E'], fontsize=20)
plt.yticks(range(20,81,10), ["[20]","[30]","[40]","[50]","[60]","[70]","[80]"], 
          fontsize=20, color="red")
```

### legend

```python
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend()
```

```python
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange")
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey")
plt.legend(['A','B'], loc="center") # label을 legend 함수에 입력할 수 있음
```

```python
# 범례 위치 바꾸기 (1)
plt.subplot(2,2,1)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="upper left")

plt.subplot(2,2,2)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="upper right")

plt.subplot(2,2,3)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="lower left")

plt.subplot(2,2,4)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="lower right")
```

```python
# 범례 위치 바꾸기 (2)
plt.subplot(2,2,1)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="center right")

plt.subplot(2,2,2)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="center left")

plt.subplot(2,2,3)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="upper center")

plt.subplot(2,2,4)
plt.plot([-2,-1,0,1,2],[1,3,9,8,4], marker='^', color="orange", label='A')
plt.plot([-2,-1,0,1,2],[10,7,3,5,2], marker='>', color="grey", label='B')
plt.legend(loc="lower center")
```

### seaborn

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.plot(np.random.normal(0,100,1000).cumsum(), color="purple")
```

```python
sns.set(style="darkgrid")
plt.plot(np.random.normal(0,100,1000).cumsum(), color="grey")
```

```python
# 정규분포를 따르는 데이터 100개를 추출 후 그래프로 표현
sns.displot(np.random.normal(0,100,100))
```

```python
import pandas as pd
```

```python
# mpg 데이터를 활용해서 그래프 그리기
mpg = pd.read_csv("mpg.csv")
sns.relplot(data=mpg, x="cty", y="displ", hue="cyl", 
            palette=["blue", "green", "magenta", "red"])
```

```python
sns.relplot(data=mpg, x="hwy", y="displ", hue="manufacturer")
```

```python
sns.boxplot(data=mpg, x="drv", y="displ")
```

```python
sns.catplot(data=mpg, x="category", y="displ")
```
