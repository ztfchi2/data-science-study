### append
- df1.append(df2) 형태로 사용
- 세로 결합

```python
df1 = df.Day.head(5)
```

```python
df2 = df.Day.tail(5)
```

```python
df1.append(df2) # df1에 세로(칼럼)방향으로 df2를 결합
```

```python
df3 = df.loc[[3,101,7,54,10,28,19]]
```

```python
df4 = df.loc[[1,10,105,77,69,112,5]]
```

```python
df3.append(df4, sort=True) # sort: True이면 칼럼을 알파벳 순서대로 정렬
```

### describe

- 기초통계량을 생성하여 간단하게 데이터를 요약함

```python
df.describe() # df의 각 칼럼별 기초통계량을 보여줌
```

```python
df.describe(percentiles=[.01,.05,.25,.75,.95,.99]) # percentiles: 사용자 지정 백분위수 추가
```

```python
df.info()
```

```python
df.describe(include=["object"]) # include: 지정한 자료형만 출력
```

```python
df.describe(exclude=["float"]) # exclude: 지정한 자료형만 제외하고 출력
```

### drop_duplicates
- df.drop_duplicates 형태로 사용
- 중복값 제거

```python
df5 = df.Cases_Guinea.tail(10)
```

```python
df5
```

```python
df5.drop_duplicates() # 중복된 값을 제거
```

```python
df5.drop_duplicates(keep="last") # keep: 어느 위치에 있는 중복값을 남길지 결정, last -> 뒤에 나오는 값만 남기고 중복값 제거
```

### equals
- df1.equals(df2) 형태로 사용
- 동일한 요소가 포함되어 있는지 확인

```python
df1.equals(df2) # df1와 df2의 내용이 다르므로 False 반환
```

```python
df1_1 = df.Day.loc[[0,1,2,3,4]]
```

```python
df1.equals(df1_1) # df1와 df1_1에 동일한 요소가 포함되어 있으므로 True 반환
```

### _get_value
- df._get_value(index, col) 형태로 사용
- 시리즈나 데이터프레임의 값을 구함 (values와 동일하게 작동)

```python
df._get_value(111, "Date") # 111행의 Date 값을 구함
```

```python
df._get_value(43, "Day") # 43행의 Day 값을 구함
```

### isin
- df.isin(values) 형태로 사용
- 특정한 값을 포함한 행들을 골라냄

```python
df2.isin([0,3]) # 0이나 3을 포함하고 있을 경우 True
```

```python
df6 = df[["Day","Cases_Liberia"]].head(12)
```

```python
df6
```

```python
df6.isin([281,288,np.NaN]) # 값들 중에 281, 288, NaN을 포함하고 있을 경우 True
```

### replace
- df.replace(a, b) 형태로 사용
- 특정 값을 다른 값으로 치환

```python
df2.replace(4,0) # 4를 0으로 바꿈
```

```python
df2.replace([1,2,3,4], 0) # 1,2,3,4를 0으로 바꿈
```

### sort_values
- df.sort_values(by=col) 형태로 사용
- 칼럼을 오름차순/내림차순 정렬

```python
df4.sort_values(by="Day") # Day 칼럼을 기준으로 오름차순 정렬
```

```python
df4.sort_values(by="Day", ascending=False) # Day 칼럼을 기준으로 내림차순 정렬
```
