### 시리즈 추가 다루기

```python
mpg_cty = mpg["cty"]
```

```python
mpg_cty[mpg_cty > mpg_cty.mean()] # mpg_cty값 중 평균 이상인 값을 불러와라!
```

```python
mpg_cty + mpg_cty # 두 개의 값이 더해진다.
```

```python
mpg_hwy = mpg["hwy"]
```

```python
mpg_hwy + mpg_cty
```

```python
mpg_hwy * mpg_cty # 사칙연산 다 가능하다!
```

```python
# 인덱스도 정렬을 할 수 있다.
mpg_hwy.sort_index(ascending=False) # 내림차순으로 정렬할 수 있다.
```

```python
mpg_hwy.sort_index(ascending=True) # 오름차순으로 정렬할 수 있다.
```

### 시계열 데이터프레임 만들기
- datetime 판다스에서 제공하는 시계열 데이터 형식
- 시계열 데이터는 Dtype을 꼭 datetime으로 변경해야 한다!
- 이유는 datetime 문법이 따로 있고, 이 문법을 적용시키려면 Dtype이 꼭 datetime으로 설정되어있어야 한다.

```python
df1 = pd.read_csv("country_timeseries.csv")
```

```python
df1.info() # 문자열 데이터
```

```python
df1
```

```python
df1["Date"]
```

```python
df1[df1["Date"].str.contains("2015")] # Date는 문자열 데이터
```

```python
df1["Date"] = pd.to_datetime(df1["Date"]) # datetime으로 변경해서 시계열 형식으로 맞춤
```
