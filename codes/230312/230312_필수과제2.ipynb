### 정확도(Accuracy)
- 전체 test 데이터 중 맞게 예측한 데이터의 수(비율)
- 가장 직관적인 성능 평가 지표
- 0 ~ 1 사이의  값을 가지고, 높을수록 좋다.

#### 정확도 공식
$$y=\frac{TP + TN}{Total}$$

### 정밀도(Precision)
- 양성이라고 판단한 데이터 중 실제 양성인 데이터의 수(비율)
- 예측한 값이 얼마나 정확한지 파악하는 지표
- 0 ~ 1 사이의  값을 가지고, 높을수록 좋다.

#### 정밀도 공식
$$y=\frac{TP}{TP + FP}$$

### 재현율(Recall)
- 실제 양성인 데이터 중 양성으로 예측된 데이터의 수(비율)
- 정답을 얼마나 잘 맞췄는지 파악하는 지표
- 0 ~ 1 사이의 값을 가지고, 높을수록 좋다.

#### 재현율 공식
$$y=\frac{TP}{TP + FN}$$

### F1 score
- 정밀도(Precision)와 재현율(Recall)의 조화평균
- 그룹별 데이터 불균형이 발생할 시 사용한다.
- 0 ~ 1 사이의 값을 가지고, 높을수록 좋다.

#### F1 score 공식
$$y=\frac{2}{\frac{1}{Precision} + \frac{1}{Recall}}=\frac{TP}{TP + \frac{1}{2}(FP + FN)}$$

### ROC-curve
- Receiver Operating Characteristic curve
- Recall와 Fallout의 변화를 시각화한 곡선으로, 각각을 축으로 하는 그래프를 해석해서 성능을 파악한다.
- 왼쪽 위에 가까울수록 성능이 좋다고 평가할 수 있다.
