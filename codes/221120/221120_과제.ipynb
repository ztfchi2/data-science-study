### 필수과제
- 기상청 데이터에서 날씨 데이터를 받아오셔서 melt 함수를 이용해서 깔끔하게 정리해 주세요!
- 22년 10, 11월 / 21년 10, 11월 두 개를 비교해서 날씨에 대한 비교도 간단하게 부탁드릴게요!

```python
import pandas as pd
```

```python
# 2021, 2022년 10-11월 날씨 데이터 읽기
w21 = pd.read_csv("2021_10-11.csv")
w22 = pd.read_csv("2022_10-11.csv")
```

- 공통적으로 강수량, 일조량 데이터에 결측값이 있다.

### melt 함수 이용

```python
# melt함수로 날씨 데이터를 하나로 묶음
w21_1 = pd.melt(w21, id_vars=["date"], var_name="weather")
```

```python
w21_1
```

```python
w22_1 = pd.melt(w22, id_vars=["date"], var_name="weather")
```

```python
w22_1
```

### 2021년/2022년 비교

```python
# 2021년
w21_1.groupby("weather").agg(mean=("value","mean"), 
                             median=("value","median"), 
                             max=("value","max"),
                             min=("value","min"))
```

```python
# 2022년
w22_1.groupby("weather").agg(mean=("value","mean"), 
                             median=("value","median"), 
                             max=("value","max"),
                             min=("value","min"))
```

- 2021년의 평균/중앙값 기온과 지면온도가 2022년에 비해 1도 정도 낮다.
- 그러나 이런 차이는 2022년 11월 데이터가 일부 포함되지 않아서 발생한 것으로 보인다.
- 2021년: 10/1~11/30
- 2022년: 10/1~11/23
- 나머지 날씨 정보에는 연도별 차이가 크게 보이지 않는다.
