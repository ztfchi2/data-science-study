### groupby, agg
- 그룹화, 그룹 통계치 구하기

```python
mpg_1 = mpg.groupby("manufacturer").agg(mean_cty=("cty","mean"), 
                                        mean_hwy=("hwy","mean"))
```

```python
mpg_1
```

```python
mpg_1.reset_index()
```

```python
mpg.groupby("manufacturer", as_index=False).agg(mean_cty=("cty","mean"), 
                               mean_hwy=("hwy","mean"))
```

```python
mpg_1.columns
```

```python
mpg_1.reset_index().columns
```

```python
mpg_2 = mpg.groupby("category").agg(max_displ=("displ","max"), 
                                   min_displ=("displ","min"))
```

```python
mpg_2
```

```python
mpg_2.columns
```

```python
mpg_2.reset_index().columns
```

```python
mpg_3 = mpg.assign(diff=mpg["hwy"]-mpg["cty"])
```

```python
mpg_3
```

```python
mpg_3 = mpg_3.groupby(["manufacturer","model"]).agg(median_diff=("diff","median"), 
                                                 std_diff=("diff","std"), 
                                                 cnt=("diff","count"))
```

```python
mpg_3
```

```python
mpg_3.columns
```

```python
mpg_3.reset_index().columns
```

```python
mpg.groupby(["manufacturer","cyl"]).agg(mean_displ=("displ","mean"), 
                                       min_cty=("cty","min"), 
                                       min_hwy=("hwy","min"))
```

```python
mpg.query("manufacturer == 'chevrolet'").groupby("cyl").agg(cnt=("year","count"))
```

### merge, concat
- 데이터 병합

```python
df1 = mpg.query("manufacturer == 'audi' and model == 'a4'")[["cty","hwy"]]
```

```python
df2 = mpg.query("manufacturer == 'audi' and model == 'a4'")[["cty","displ"]]
```

```python
display(df1)
display(df2)
```

```python
pd.concat([df1,df2], axis=0)
```

```python
pd.concat([df1,df2], axis=1)
```

```python
df3 = mpg.query("manufacturer == 'hyundai' and model == 'sonata'")[["cty","hwy"]]
```

```python
df4 = mpg.query("manufacturer == 'hyundai' and model == 'sonata'")[["cty","displ"]]
```

```python
display(df3)
display(df4)
```

```python
pd.concat([df1,df3], axis=0)
```

```python
pd.concat([df1,df3], axis=1)
```

```python
pd.concat([df1,df4], axis=0)
```

```python
pd.concat([df1,df4], axis=1)
```

```python
df1 = df1.reset_index()
df2 = df2.reset_index()
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on="index")
```

```python
pd.merge(df1, df2, on=["index","cty"])
```

```python
df3 = df3.reset_index()
df4 = df4.reset_index()
```

```python
display(df3)
display(df4)
```

```python
pd.merge(df3, df4, on="index")
```

```python
pd.merge(df3, df4, on=["index","cty"])
```
