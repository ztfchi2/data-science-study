### groupby, agg

```python
import pandas as pd
```

```python
exam = pd.read_csv("exam.csv")
```

```python
df_1 = exam.groupby("nclass").agg(mean_math=("math", "mean"))
```

```python
df_1
```

```python
df_1.columns
```

```python
df_1.reset_index().columns
```

```python
exam.groupby("nclass", as_index=True).agg(mean_math=("math", "mean"))
```

```python
exam.groupby("nclass", as_index=False).agg(mean_math=("math", "mean"))
```

```python
df_1.reset_index() # 인덱스를 컬럼으로 바꾸기!
```

### 데이터 병합
- merge, concat (join을 이용해서!)

```python
df1 = pd.DataFrame({'A':[1,2,3], 
                   'B':[4,5,6]})
df2 = pd.DataFrame({'A':[1,2,3], 
                   'C':[7,8,9]})
```

```python
display(df1)
display(df2)
```

```python
# 병합 진행 (행)
pd.concat([df1,df2], axis=0)
```

```python
# 병합 진행 (열)
pd.concat([df1,df2], axis=1)
```

- merge

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how="left", on="A")
# on: 교집합의 개념
```

### 교집합이 2개 이상인 경우!

```python
df1 = pd.DataFrame({'A':[1,2,3], 
                   'B':[4,5,6], 
                   'C':[13,14,15]})
df2 = pd.DataFrame({'A':[1,2,3], 
                   'B':[4,5,7], 
                   'D':[10,11,12]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how="inner", on=["A","B"]) # 정말 교집합만 반환. A, B를 기준으로 모두 다 같아야 한다!
```

- display 함수를 사용하는 이유

```python
df1
df2
```

```python
display(df1)
display(df2)
```

```python
print(df1)
print(df2)
```

### 평균이 아닌 0으로 채운다

```python
exam.loc[[3,7,12],["science"]] = np.nan
```

```python
exam["science"] = exam["science"].fillna(0)
```

```python
exam["science"].mean() # na값으로 0으로 대체 시
```

```python
exam.loc[[3,7,12],["science"]] = np.nan
exam["science"] = exam["science"].fillna((4+5)/3) # 안에 숫자가 나오는 공식도 사용할 수 있다.
```
