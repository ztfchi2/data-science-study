```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

### 의사결정나무 DT에 대해서 구체적으로 알아보자!
- 의사결정나무는 트리가 깊어지면서 과적합이 일어날 확률이 높다.
- 실질적으로 크게 많이 사용하는 모델은 아니다.
- 하지만 모델링이 아니어도 인사이트를 위해서 한번쯤 살펴보면 좋은 것들이 있다.
- DT 쉽다, 직관적이다. (분석하기 전에 전반적인 데이터를 이해하기도 쉽다.)
- 모델링이 예측에 대한 평가지표를 올리는 것도 목적이지만, 모델링을 통해 피처의 중요도 또는 시각화를 보고 가설에 대한 해설 또는 추가적인 분석으로 접근할 수 있다. 메인 분석 전에 가볍게 살펴보는 정도?
- 단점 : 과적합이 일어난다.
- 왜 과적합이 일어나고 과적합을 줄이기 위해서는 어떤 식으로 조절해야 하는가?
- 알고리즘 성능도 떨어집니다.

```python
dt_clf = DecisionTreeClassifier(random_state=156)

# iris 데이터 가지고 확인해 보기
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, 
                                                    test_size=0.3, random_state=156)

dt_clf = dt_clf.fit(X_train, y_train)
```

### 과적합을 피하기 위해서는?
- 하이퍼파라미터가 존재한다.
- 하이퍼파라미터를 통해서 가지치기나, 최대 트리 깊이나, ccp나, weight, criterion 분할할 gini, entropy, log_loss를 선정할 수 있다.

- 하이퍼파라미터에 무엇이 있나?
- criterion : 지니, 엔트로피, 로그 등으로 기준을 정할 수 있고, 디폴트는 지니
- splitter : best, random, best 디폴트 값
- max_depth : 트리의 최대 깊이를 규정, 과적합을 조절/제어 할 수 있음
- min_samples_split : 분할하기 위한 최소한의 샘플 데이터 수
- min_samples_leaf : 밑단에 대한 로드가 되기 위한 최소한의 샘플 데이터 수
- max_features : 고려해야 할 최대 피처 개수, auto, sqrt(전체 피처 개수의 루트만큼), log(전체 피처 개수에서 log2), None(전체 피처 선정)
- min_impurity_decrease : 디폴트는 0, 원하는 수만큼 지정할 수 있다.
- class_weight : 이진 분류가 아니라 다중 분류인 경우는 class별로 가중치를 부여할 수 있다. ex) (0, 1, 2) -> [{0:1, 1:2, 2:3}] 가중치를 줄 수 있다.
- ccp_alpha : Minimal Cost - Complexity pruning, 해당 수식이 존재한다. 

### 실제 과적합을 막기 위해 하이퍼파라미터를 튜닝한 경우

### 예측을 해서 예측 정확도도 체크해 보자!

```python
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier(random_state=156)

# iris 데이터 가지고 확인해 보기
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, 
                                                    test_size=0.3, random_state=156)

dt_clf = dt_clf.fit(X_train, y_train)
```

```python
# 하이퍼파라미터를 튜닝하지 않은 경우
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print("하이퍼파라미터 추가 전 기본 베이스", accuracy)
print(dt_clf.get_params())
```

```python
dt_clf = DecisionTreeClassifier(max_depth=4, min_samples_split=7, random_state=156)

# iris 데이터 가지고 확인해 보기
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, 
                                                    test_size=0.3, random_state=156)

dt_clf = dt_clf.fit(X_train, y_train)
```

```python
# 하이퍼파라미터를 튜닝한 경우
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print("하이퍼파라미터 추가 전 기본 베이스", accuracy)
print(dt_clf.get_params())
```

- 튜닝하기 전에 비해 튜닝 후의 정확도가 떨어졌다.
- 이게 의미하는 것은 무엇인가?
- 과적합에 대한 것을 바로바로 확인할 수 있고, 정확도가 높다고 해서 다 좋은 것은 아니다. 분석가는 꼭 과적합에 대한 판단을 해야 한다.
