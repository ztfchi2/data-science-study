```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris

import numpy as np
```

### 저번시간에 진행한 kfold skfold 더 쉽게 하는 방법
- 사이킷런에서 제공하고 있는 cross_val_score 교차검증 가능
- cross_val_score(사용할 모델, x(데이터), y(데이터), scoring(평가할 스코어링 방법), cv(몇 번 교차검증 할 것인가?), n_jobs(cpu 사용에 대한 지정, -1이면 모든 cpu를 사용), verbose(진행상황에 대한 유무 보기), fit_params 등등)
- 대부분 사용하는 건 내가 사용할 모델, 데이터, 평가점수, 몇 번 교차검증?
- cross_val_score 자동적으로 분류 or 회귀에 따라서 kfold, skfold 변경된다.
- 우리가 모델을 입력할 때 Classifier 자동으로 skf / 회귀 진행 시 KFold로 바뀐다.

- 대부분의 사이킷런 제공은 이진 분류에서 많이 사용되고, 특히 평가지표들은 대부분 이진분류로 진행된다.

```python
iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=111)

data = iris_data.data
label = iris_data.target
```

```python
# 교차검증 진행 (accuracy)
scores = cross_val_score(dt_clf, data, label, scoring="accuracy", cv=10)
print(np.round(scores, 4))
print(np.round(np.mean(scores), 4))
```

```python
# 교차검증 진행 (precision)
scores = cross_val_score(dt_clf, data, label, scoring="precision_macro", cv=10)
print(np.round(scores, 4))
print(np.round(np.mean(scores), 4))
```

```python
# 교차검증 진행 (recall)
scores = cross_val_score(dt_clf, data, label, scoring="recall_macro", cv=10)
print(np.round(scores, 4))
print(np.round(np.mean(scores), 4))
```

```python
# 교차검증 진행 (f1 score)
scores = cross_val_score(dt_clf, data, label, scoring="f1_macro", cv=10)
print(np.round(scores, 4))
print(np.round(np.mean(scores), 4))
```

### 평가지표 (Evaluation)
- 정확도, 정밀도, 재현율, f1 평가지표
- 성능평가지표는 모델링에 대한 성능 지표, 분류냐 회귀냐에 따라 달라진다.
- 회귀는 실제값과 예측값에 대한 잔차, 오차에 대한 기준으로 스코어링한다.
- 오차값에 절댓값 쓰위 평균 오차를 구하고 RMSE, MSE

- 우리가 배웠던 분류 평가 방법이 4-5개 정도 된다.
- 평가지표는
- 정확도(accuracy)
- 정밀도(precision)
- 재현율(recall)
- f1 스코어
- ROC AUC
- 오차행렬에 대한 내용
- 위의 평가지표는 이진분류 시 사용된다.

### 오차 행렬 confusion matrix
- 4 사분면에 positive 1 / negative 0

        - 예측 클래스
        - N(0) / P(1)
- N (0)
        - TN FP
- 실제 클래스
- P (1)

        - FN TP
- TN은 뭐냐?
- True 실제값과, 예측값의 클래스 같다 True + N 예측값에 대한 N = TN

- TN 예측값은 Neg(0), 실제값 역시 Neg(0) 값
- 이 원리를 그대로 적용해서 FP, FN, TP가 나오는 것

### 정확도
- 예측 결과가 동일한 데이터 개수 / 전체 예측 데이터 개수
- 정확도 계산식
- (TN + TP) / (TN + TP + FN + FP)
- 절대로 정확도만 가지고 평가할 수 없다.
- 정확도 지표만 보면 안 되는 이유?
- 100개 데이터가 있다. 90개는 0, 10개는 1, 0을 예측하는 게 90% 클래스 불균형
- 오차행렬을 만드는 이유가 정확도만 보면 문제가 되어서 진행하는 것이다.

### 정밀도
- TP / (FP + TP)
- 1로 예측한 것들 중 진짜 1인 경우가 얼마나 되는가?
- 예측을 Positive로 한 대상 중에서 실제 값이 Positive로 일치한 데이터
- positive에 대한 성능을 정밀하게 예측한다고 해서 '양성 예측도'라고도 한다.

### 재현율
- 실제 값이 Positive 대상 중에서 예측과 실제 값이 Positive로 일치한 데이터 비율
- TP / (FN + TP)
- 민감도(Sensitivity), TPR(True Positive Rate)라고도 한다.

### 정밀도와 재현율은 Trade off 관계
- 정밀도 - TP가 올라가거나, FP 점수
- TP / (FP + TP)

- 재현율 - FN 점수
- TP / (FN + TP)

- 임계값은 1과 0에 대해서 예측할 때 디폴트 값이 0.5를 기준으로 해서 1이 되거나 0.499 0이 되거나
- 분류를 하는데, 해당 임계값이 조절 가능하다.
- 0.6부터 1로, 0.59 아래는 0으로 조절이 가능하다.
